```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fundos.csv')
```

```python
df = df.fillna(0)
```

```python
df['Liquidez Diária'] = df['Liquidez Diária'].astype(int)
```

```python
id_liq = df[df['Liquidez Diária'] < 5000].index
df.drop(id_liq, inplace=True )

quant = df[df['Quant. Ativos'] == 1].index
df.drop(quant, inplace = True)

id_dy = df[df['DY(6m)'] == '0,00%'].index
df.drop(id_dy, inplace=True)
```

```python
sh = df[np.logical_and(df['Setor'] == 'Shoppings', df['Quant. Ativos']<5)].index
df.drop(sh, inplace = True)
```

```python
df['Preço Atual'] = df['Preço Atual'].map(lambda x: x.lstrip('R$'))
df['Dividendo'] = df['Dividendo'].map(lambda x: x.lstrip('R$'))
df['Patr. Líq.'] = df['Patr. Líq.'].map(lambda x: x.lstrip('R$'))
df['VPA'] = df['VPA'].map(lambda x: x.lstrip('R$'))
```

```python
#df['Preço Atual'] = df['Preço Atual'].apply(str)
df['Preço Atual'] = df['Preço Atual'].replace(regex=[','], value='.')
df['Dividendo'] = df['Dividendo'].replace(regex=[','], value='.')
df['Preço Atual'] = df['Preço Atual'].astype(float)
df['Dividendo'] = df['Dividendo'].astype(float)
```
