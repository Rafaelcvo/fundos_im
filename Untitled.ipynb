```python
import numpy as np
import pandas as pd
```

```python
#exibindo todas as colunas

pd.options.display.max_columns = 28
pd.options.display.max_rows = 138
```

```python
df = pd.read_csv('fundos.csv')
```

```python
df = df.fillna('0')
```

```python
df['Liquidez Diária'] = df['Liquidez Diária'].astype(int)
```

```python
df['Preço Atual'] = df['Preço Atual'].str.strip('R$')
df['Preço Atual'] = df['Preço Atual'].str.replace('.','')
df['Preço Atual'] = df['Preço Atual'].str.replace(',','.')
df['Preço Atual'] = df['Preço Atual'].astype(float)

df['Dividendo'] = df['Dividendo'].str.strip('R$')
df['Patr. Líq.'] = df['Patr. Líq.'].str.strip('R$')
df['VPA'] = df['VPA'].str.strip('R$')

df['Dividendo'] = df['Dividendo'].replace(regex=[','], value='.')
df['Dividendo'] = df['Dividendo'].astype(float)

df['P/VPA'] = df['P/VPA'].replace(regex=[','],value='.')
df['P/VPA'] = df['P/VPA'].astype(float)
```

```python
df['DY Ano'] = df['DY Ano'].str.strip('%')
df['DY Ano'] = df['DY Ano'].replace(regex=[','], value='.')
df['DY Ano'] = df['DY Ano'].astype(float)
```

```python
#removendo fundos que nao pagaram dividendos durante o ano.
rem_div = df[df['DY Ano'] == 0].index
df.drop(rem_div, inplace=True)
```

```python
#removendo fundos com liquidez menor que 100
rem_liq = df[df['Liquidez Diária'] < 100].index
df.drop(rem_liq, inplace=True)
```

```python
#removendo P/VPA maior que 100%
rem_vpa = df[df['P/VPA'] > 1].index
df.drop(rem_vpa, inplace=True)
```

```python
#eliminando shoppings igual a 1
rem_shop = df[np.logical_and(df['Setor'] == 'Shoppings',df['Quant. Ativos'] == 1)].index
df.drop(rem_shop, inplace=True)
```

```python
#eliminando lajes igual a 1
rem_lajes = df[np.logical_and(df['Setor'] == 'Lajes Corporativas',df['Quant. Ativos'] == 1)].index
df.drop(rem_lajes, inplace=True)
```

```python
#simulando retorno mensal investindo 20k
df['Quantidade'] = 20000 / df['Preço Atual']
df['Quantidade'] = round(df['Quantidade'], 2)
df['Ret 20k'] = df['Quantidade'] * df['Dividendo']
df['Ret 20k'] = round(df['Ret 20k'],2)
```

```python
cp = df.copy()
```
